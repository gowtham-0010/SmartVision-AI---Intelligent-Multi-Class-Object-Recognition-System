```python
from ultralytics import YOLO
import os
```

```python
print("Train Images:", len(os.listdir("../data/detection/images/train")))
print("Val Images:", len(os.listdir("../data/detection/images/val")))
```

```python
yaml_content = """
path: ../data/detection

train: images/train
val: images/val

nc: 25

names:
  0: person
  1: bicycle
  2: car
  3: motorcycle
  4: airplane
  5: bus
  6: truck
  7: traffic light
  8: stop sign
  9: bench
  10: bird
  11: cat
  12: dog
  13: horse
  14: cow
  15: elephant
  16: bottle
  17: cup
  18: bowl
  19: pizza
  20: cake
  21: chair
  22: couch
  23: bed
  24: potted plant
"""

with open("../data/detection/data.yaml", "w") as f:
    f.write(yaml_content)

print("data.yaml created ✅")
```

```python
model = YOLO("yolov8n.pt")
```

```python
model.train(
    data="../data/detection/data.yaml",
    epochs=10,        
    imgsz=416,        
    batch=4,          
    name="smartvision_yolo_fast"
)
```

```python
metrics = model.val()
print(metrics)
```

```python
model.predict(
    source="../data/detection/images/val",
    save=True,
    conf=0.4
)
```

```python
from ultralytics import YOLO

best_model = YOLO("../runs/detect/smartvision_yolo_fast/weights/best.pt")
```

```python
best_model.predict(
    source="../data/detection/images/val",
    show=True,
    conf=0.5
)
```
